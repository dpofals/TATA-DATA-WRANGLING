```
def clean_data(data):
    cleaned_data = []
    for item in data:
        Quantity = item[0]
        UnitPrice = item[1]
        if check_valid_inputs(Quantity, UnitPrice):
            cleaned_data.append(item)
    return cleaned_data
```
